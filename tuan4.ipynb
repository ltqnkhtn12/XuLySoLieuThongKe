```python
import numpy as np

# Bước 1: Load dữ liệu từ
data = pd.DataFrame(data)

# Bước 2: Tính giá trị trung bình cho từng cột và ghép lại thành vecto trung bình µ
mean_vector = np.mean(data, axis=0)

# Bước 3: Tính ma trận B = X - µ1.T (với µ1 là vecto trung bình mở rộng)
h = np.ones((data.shape[0], 1))
B = data - np.dot(h, mean_vector[np.newaxis, :])

# Bước 4: Tính ma trận covariance C = BTB
C = np.cov(B, rowvar=False)

# Bước 5: Tìm giá trị riêng và vecto riêng
eigenvalues, eigenvectors = np.linalg.eig(C)

# Bước 6: Sắp xếp giá trị riêng và vecto riêng theo thứ tự giảm dần
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Bước 7: Tính tỷ lệ phần trăm tích luỹ
cumulative_var_ratio = np.cumsum(eigenvalues) / np.sum(eigenvalues)

# Bước 8: Chọn số thành phần chính dựa trên tỷ lệ tích luỹ
threshold = 0.9
num_components = np.argmax(cumulative_var_ratio >= threshold) + 1

# Bước 9: Trích xuất ma trận W gồm các vecto riêng tương ứng
W = eigenvectors[:, :num_components]

# Bước 10: Dữ liệu trên cơ sở mới với số chiều đã giảm
T = np.dot(B, W)

# In số chiều sau giảm
print(f"Số chiều sau khi giảm: {num_components}")

# In tỷ lệ phần trăm tích luỹ
print(f"Tỷ lệ phần trăm tích luỹ: {cumulative_var_ratio[num_components - 1] * 100:.2f}%")
```

```python
from sklearn.decomposition import PCA

# Khởi tạo mô hình PCA với số thành phần chính mong muốn (trong trường hợp này, chọn 2 thành phần chính)
pca = PCA(n_components=2)

# Thực hiện phân tích PCA trên dữ liệu
principal_components = pca.fit_transform(data)

# Tạo DataFrame mới từ các thành phần chính
df_reduced = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

print(df_reduced)
```
